```python
import pandas as pd 
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook"
config = dict({'scrollZoom': True})
```

```python
data = pd.read_csv('../Data/InternationalCovid19Cases.csv')
data
```

```python
case_data = data[['iso_code', 'date', 'new_cases', 'total_cases', 'new_deaths', 'total_deaths']]
case_data = case_data.rename(columns={'iso_code': 'Code', 'date': 'Date', 'new_cases': 'New Cases', 'new_deaths': 'New Deaths', 'total_cases': 'Total Cases', 'total_deaths': 'Total Deaths'})
case_data = case_data.fillna(0)
```

```python
case_data
```

```python
can = case_data[case_data['Code'] == 'CAN']
usa = case_data[case_data['Code'] == 'USA']

can.insert(0, 'Country', ['Canada']*551, True)
usa.insert(0, 'Country', ['United States']*555, True)

can = can.drop(columns='Code')
usa = usa.drop(columns='Code')
```

```python
can.insert(2, 'New Cases/Total', can['New Cases']/can['Total Cases'], True)
usa.insert(2, 'New Cases/Total', usa['New Cases']/usa['Total Cases'], True)
```

```python
can.insert(5, 'New Deaths/Total', can['New Deaths']/can['Total Deaths'], True)
usa.insert(5, 'New Deaths/Total', usa['New Deaths']/usa['Total Deaths'], True)
```

```python
can
```

```python
usa
```

```python
combined = pd.concat([usa,can])
```

```python
fig = px.line(combined, x='Date', y='Total Deaths', color='Country', template = "plotly_dark", title = "New Cases Over Time in the USA vs Canada")

fig.show(config=config)
```
